```python
import re
def extract_num(x) :
#     x = ''.join(re.findall('([+-]?[0-9]*[.]?[0-9]*)', str(x)))
    x = re.sub('[(].+[)]', '', x)
    x= ''.join(re.findall('[-+]?\d*\.\d+|\d+', str(x)))
    return x

def my_strip_split (df) :
    for column in ['보호자명', '환자명', '생일', '검사일', '병원', '표준항목'] :
        df[column] = df[column].apply(lambda x : str(x).strip())
    for column in ['생일', '검사일'] :
        df[column] = df[column].apply(lambda x : x.split(' ')[0])
    for column in ['병원'] :
        df[column] = df[column].apply(lambda x : x.split('_')[0])
    for column in ['검사결과'] :
#         df[column] = df[column].apply(lambda x : re.findall(r"[-+]?\d*\.\d+|\d+", x))
#         df[column] = df[column].apply(lambda x : re.sub('[\[\]]', '', (x)))
#         df[column] = df[column].astype('float64')
#         df[column] = df[column].apply(lambda x : re.sub('[^0-9.+-]', '', x))
#         df[column] = df[column].apply(lambda x : remove_bar(x))
        # extract digit only
        df[column] = df[column].apply(lambda x : extract_num(x))
        
    return df

# data preporcessing
df2 = my_strip_split(df2)
df3 = my_strip_split(df3)    

df2.to_csv('result/df02_1.csv', index=None)
df3.to_csv('result/df03_1.csv', index=None)
```

```python
check_list = ['보호자명', '환자명', '생일', '검사일', '병원']
def make_newdf(df, labeled_df, column_list) :
    count = -1
    print(df.shape)
    print(labeled_df.shape)
    new_list = check_list + column_list
    df_new = pd.DataFrame(columns = new_list)
    df_new['진단코드'] = ''
    print(df_new.columns)
    par, pat, bir, tes, hos = '','','','',''
   
    for i in range(len(df)) :
        
        var_list = []
        for j in range(len(check_list)) :
            var_list.append(df.loc[i,check_list[j]])
            
        if ((par != var_list[0]) & (pat != var_list[1]) & (bir != var_list[2]) 
           & (tes != var_list[3])) or (hos != var_list[4]):
            count += 1
            for j in range(len(check_list)) :
                df_new.loc[count, check_list[j]] = var_list[j]
#             df_new.loc[count, '보호자명'] = var_list[0]
#             df_new.loc[count, '환자명'] = var_list[1]
#             df_new.loc[count, '생일'] = var_list[2]
#             df_new.loc[count, '검사일'] = var_list[3]   
#             df_new.loc[count, '병원'] = var_list[4]
            
            temp = labeled_df.loc[(labeled_df['보호자명']==var_list[0]) & (labeled_df['환자명']==var_list[1]) & \
                         (labeled_df['생일']==var_list[2]) & (labeled_df['검사일']==var_list[3]) & \
                         (labeled_df['병원']==var_list[4])]
            if len(temp) > 0 :
                df_new.loc[count, '진단코드'] = temp.loc[temp.index.values[0], '진단코드']
            
        else :
            item = str(df.loc[i, '표준항목']).strip()
#             print(item)
            if item in df_new.columns :
                df_new.loc[count, item] = df.loc[i, '검사결과']   
            
        par, pat, bir, tes, hos = var_list[0], var_list[1], var_list[2], var_list[3], var_list[4]
        
    return df_new
```

```python
def regex_label( df) :
    label = '진단코드'

    df[label] = df[label].apply(lambda x : re.sub('중성화수술','Neutral', str(x)))
    df[label] = df[label].apply(lambda x : re.sub('중성화 수술','Neutral', str(x)))
    df[label] = df[label].apply(lambda x : re.sub('중성화','Neutral', str(x)))
    df[label] = df[label].apply(lambda x : re.sub('여아중성화','Neutral', str(x)))
    df[label] = df[label].apply(lambda x : re.sub('남아중성화','Neutral', str(x)))
    df[label] = df[label].apply(lambda x : re.sub('여아Neutral','Neutral', str(x)))
    df[label] = df[label].apply(lambda x : re.sub('남아Neutral','Neutral', str(x)))
    df[label] = df[label].apply(lambda x : re.sub('여중','Neutral', str(x)))
    df[label] = df[label].apply(lambda x : re.sub('남중','Neutral', str(x)))


    df[label] = df[label].apply(lambda x : re.sub('[건강][가-힣 ()a-zA-Z,0-9\/]+','test', str(x)))
    # df.label = df.label.apply(lambda x : re.sub('건강검진(Castration)','test', str(x))) '건강검진(Spay)'

    df[label] = df[label].apply(lambda x : re.sub('X','normal', str(x)))
    df[label] = df[label].apply(lambda x : re.sub('x','normal', str(x)))
    df[label] = df[label].apply(lambda x : re.sub('이상없음','normal', str(x)))
    df[label] = df[label].apply(lambda x : re.sub('XX','normal', str(x)))
    df[label] = df[label].apply(lambda x : re.sub('-','normal', str(x)))
    df[label] = df[label].apply(lambda x : re.sub('normalnormal','normal', str(x)))
    df[label] = df[label].apply(lambda x : re.sub('정상','normal', str(x)))

    # df[label] = df[label].apply(lambda x : re.sub('2170.0','2170', str(x)))
    # df[label] = df[label].apply(lambda x : re.sub('2071.0','2071', str(x)))
    df[label] = df[label].apply(lambda x : re.sub('[.]0','', str(x)))
    df[label] = df[label].apply(lambda x : re.sub('[ ]+','', str(x)))

    return df
```
